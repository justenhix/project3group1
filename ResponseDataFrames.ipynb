```python
# Dependancies
import requests
import json
from pprint import pprint
#from config import api_key
```

```python
# Calculations
import pandas as pd
import numpy as np
```

```python
# Base URL
#url = 'https://api.fiscaldata.treasury.gov/services/api/fiscal_service'
url = 'https://api.bls.gov/publicAPI/v2/timeseries/data/'
```

```python
unemp_char = 'LAUIM371674000000004'
unemp_rate_char = 'LAUIM371674000000003'
unemp_ral = 'LAUMT373958000000004'
unemp_rate_ral ='LAUMT373958000000003'
unemp_green = 'LAUMT372466000000004'
unemp_rate_green = 'LAUMT372466000000003'
unemp_win = 'LAUMT374918000000004'
unemp_rate_win = 'LAUMT374918000000003'

start_year = 2015
end_year = 2024

nc_query_url = url + unemp_char + '?startyear=' + str(start_year) + '&endyear=' + str(end_year) + '&catalog=false'


print(nc_query_url)
```

```python
nc_response = requests.get(nc_query_url)
unemp_char_response_json = nc_response.json()
```

```python
pprint (unemp_char_response_json)
```

```python
unemp_char_df = pd.DataFrame(nc_response_json["Results"]["series"][0]["data"])
unemp_char_df.head()
```

```python
columns_to_drop = ['latest', 'footnotes']
unemp_char_df = unemp_char_df.drop(columns=columns_to_drop)
print(unemp_char_df)
```
